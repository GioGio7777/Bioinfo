### Import libraries

```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

## Search Target protein

### Target search for coronavirus

```python
# Target search for coronavirus
target = new_client.target
target_query = target.search('coronavirus')
```

```python
target_query # this is dict and we use from_dict 
```

```python
targets = pd.DataFrame.from_dict(target_query)
targets
```

### Select and retrivie bioactivity data for SARS coronavirus 3C- like proteinase(fifth entry)

```python
selected_target = targets.target_chembl_id[4]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id = selected_target).filter(type = "IC50")
```

```python
res
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df
```

```python
df.type.unique()
```

#### Save data to CSV file 

```python
df.to_csv("bioactivity_data.csv",index=False)
```

## Handle missing data

```python
df2 = df[df.standard_value.notna()]
df2
```

## Data pre-process

### Labeling compounds as either being active,inactive or intermediate

```python
bioactivity_class = []
for i in df2.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append("inactive")
        
    elif float(i) >= 1000:
        bioactivity_class.append("active")
    
    else:
        bioactivity_class.append("intermediate")
```

### Alternative method

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2[selection]
```

```python
df3
```

```python
df3=pd.concat([df3,pd.Series(bioactivity_class,name="class")],axis = 1)
```

```python
df3
```

```python
df3.to_csv('bioactivity_preprocessed_data.csv',index=False)
```
