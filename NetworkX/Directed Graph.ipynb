```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
G.add_edges_from([ ("A","B"),("A","C"),("C","B"),('D',"A"),("E","D")])
```

```python
val_map = {
    "A":1.00,
    "D":0.00 
}
```

```python
values = [val_map.get(nodes, 0.50) for nodes in G.nodes()]
```

```python
pos = nx.spring_layout(G)
```

```python
nx.draw_networkx_nodes(G,pos, cmap = plt.get_cmap('Greens') , node_size = 500,node_color = values)
nx.draw_networkx_edges(G,pos, edgelist = G.edges(),edge_color = "black" ) 
nx.draw_networkx_labels(G,pos)
plt.show()
```

```python
G = nx.DiGraph()
G.add_weighted_edges_from([ ("0","1",4),("0","2",3),("1","2",3),('1',"3",4),("2","3",5)])
```

```python
val_map = {
    "0":1.00,
    "3":0.00 
}
```

```python
labels = nx.get_edge_attributes(G,"weight")
```

```python
pos = nx.planar_layout(G)
```

```python
values = [val_map.get(nodes, 0.50) for nodes in G.nodes()]
```

```python
nx.draw_networkx_nodes(G,pos, cmap = plt.get_cmap('Greens') , node_size = 500,node_color = values)
nx.draw_networkx_edges(G,pos, edgelist = G.edges(), edge_color = "black" ) 
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
nx.draw_networkx_labels(G,pos)
plt.show()
```

```python
from pygraph.classes.digraph import digraph 
from pygraph.algorithms.minmax import maximum_flow
```

```python
gr = digraph()
gr.add_nodes(["F",1,2,'B'])
gr.add_edge(("F",1,),wt=4)
gr.add_edge((1,2),wt=3)
gr.add_edge((2,"B"),wt=5)
gr.add_edge(("F",2),wt=3)
gr.add_edge((1,"B"),wt=4)
```

```python
flows,cuts = maximum_flow(gr,"F","B")
print("flow is: " , flows)
print("cut is: " , cuts)
```
